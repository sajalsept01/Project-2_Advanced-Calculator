```python
import tkinter as tk
import math

# Function to evaluate expressions
def calculate():
    try:
        expression = entry.get()
        result = eval(expression, {"__builtins__": None}, math.__dict__)
        entry.delete(0, tk.END)
        entry.insert(tk.END, str(result))
    except Exception as e:
        entry.delete(0, tk.END)
        entry.insert(tk.END, "Error")

# Function to insert text into entry
def insert_text(text):
    entry.insert(tk.END, text)

# Function to clear entry
def clear():
    entry.delete(0, tk.END)

# Create main window
root = tk.Tk()
root.title("Advanced Calculator")

# Entry widget
entry = tk.Entry(root, width=30, font=("Arial", 15))
entry.grid(row=0, column=0, columnspan=5, padx=10, pady=10)

# Buttons layout
buttons = [
    ("7",1,0), ("8",1,1), ("9",1,2), ("/",1,3), ("sqrt(",1,4),
    ("4",2,0), ("5",2,1), ("6",2,2), ("*",2,3), ("^",2,4),
    ("1",3,0), ("2",3,1), ("3",3,2), ("-",3,3), ("log(",3,4),
    ("0",4,0), (".",4,1), ("+",4,2), ("(",4,3), (")",4,4),
    ("sin(",5,0), ("cos(",5,1), ("tan(",5,2), ("log10(",5,3), ("=",5,4),
    ("asin(",6,0), ("acos(",6,1), ("atan(",6,2), ("clear",6,3)
]

# Create buttons dynamically
for (text, row, col) in buttons:
    if text == "=":
        tk.Button(root, text=text, width=8, height=2, command=calculate).grid(row=row, column=col)
    elif text == "clear":
        tk.Button(root, text=text, width=8, height=2, command=clear).grid(row=row, column=col)
    else:
        tk.Button(root, text=text, width=8, height=2, command=lambda t=text: insert_text(t)).grid(row=row, column=col)

# Run the application
root.mainloop()
```
